```python
#Practical-8
#Dhruv Thanki
#216270307058
#Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('NIFTY50_all.csv');
data['Date'] = pd.to_datetime(data['Date'])

start_date = '2000-01-03'
end_date = '2022-06-09'
date = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]

plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['Open'], label='Open Price', marker='o', linestyle='-')
plt.plot(data['Date'], data['Close'], label='Close Price', marker='o', linestyle='-')
plt.title('Stock Prices Between {} and {}'.format(start_date, end_date))
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
